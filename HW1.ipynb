```python
import codecs
import hashlib
import binascii
from binascii import hexlify, unhexlify
import re
import sys
from functools import reduce
```

```python
def xorStrings(a, b):
    return str(hex(int(a,16)^int(b,16)))[2:]
def xorArray(strings):
    return reduce((lambda x,y:xorStrings(x,y)),strings)
```

```python
def strToHex(string):
    return string.encode("utf-8").hex()
def hexToStr(hexString):
    return bytes.fromhex(hexString).decode('utf-8')
```
